## Importing Necessary Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading our data

```python
df = pd.read_csv('data/career_dataset.csv')
```

## Loading top 5 rows of data

```python
df.head()
```

## Checking out the shape of our data

```python
df.shape
```

#### Insights:

## Checking out the data types

```python
df.info()
```

#### Insights:
- Data has all categorical features

## Checking out Missing Values in Data 

```python
df.isnull().sum()
```

#### Insights: 
- Columns: Skill_Level, Tech, Leadership --> Has missing values

## Handling Missing Values

```python
for col in df.select_dtypes('object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

#### Insights:
- Missing Value handled using mode technique

## Checking for duplicates

```python
df.duplicated().sum()
```

#### Insights: 
- There is no duplicates in our data.

# <h1 style="color: cyan"> Univariate Analysis </h1>

## Checking out the Counts of Interests Based on Career

```python
plt.figure(figsize=(20, 7))
sns.countplot(data=df, x='Interest', hue='Career')
```

## Checking out the Interests Skill Level and Interests Count

```python
plt.figure(figsize=(17, 7))
plt.subplot(1,2,1)
plt.title('Interest Based On Skill Level')
sns.countplot(data=df, x='Interest', hue='Skill_Level')

plt.subplot(1,2,2)
plt.title('Interest Counts')
df['Interest'].value_counts().plot(kind='pie')
```

### Insights:
- Majority of the person has advanced skill level in their interest.
- The Interest is evenly choosen by peoples

## Checking out the Interest based on Work Style

```python
plt.figure(figsize=(15, 7))
sns.countplot(data=df, x='Interest', hue='Work_Style')
```

## Checking out the Interest based on Education

```python
plt.figure(figsize=(15, 7))
sns.countplot(data=df, x='Interest', hue='Education')
```

### Insights: 
- People prefer to do Phd more in Business and Medicine 

## Checking out the counts of each career

```python
plt.figure(figsize=(20,15))
df.Career.value_counts().plot(kind='pie')
```

### Insights: 
- Careers counts are evenly spreaded.
